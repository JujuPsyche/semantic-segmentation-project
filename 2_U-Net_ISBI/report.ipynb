## 3. 跳跃连接的作用分析
跳跃连接（Skip Connections）是 U-Net 的核心设计，其作用包括：
1. **保留空间信息**：编码器下采样会丢失细节，跳跃连接将浅层特征传递到解码器，帮助恢复高分辨率细节。
2. **缓解梯度消失**：深层网络训练时梯度容易消失，跳跃连接提供了短路路径，加速收敛。
3. **提升小目标分割**：对于神经元突触等小结构，跳跃连接能有效结合局部和全局特征。

## 4. 结论
- U-Net 在 ISBI 数据集上表现良好，验证损失稳定下降。
- 跳跃连接显著改善了边缘和小结构的分割精度。
- 后续可尝试数据增强或更深的网络结构进一步提升性能。
